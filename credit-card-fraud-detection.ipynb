## Credit Card Fraud Detection - EDA

```python
#importing libraries
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#reading data
df=pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
```

```python
#checking for null values
df.isnull().sum()
```

```python
#percentage of missing values in each row
round(100*(df.isnull().sum(axis=1)/len(df)),2).sort_values(ascending=False)
```

There is no missing / Null values either in columns or rows

```python
#descriptive statistics of data
df.describe()
```

```python
#percetage of non-frauds and frauds
print('No Frauds', round(df['Class'].value_counts()[0]/len(df)*100,2), '% of the dataset')
print('Frauds', round(df['Class'].value_counts()[1]/len(df)*100,2), '% od the dataset')
```

Most of the transcations are non-fraud,that means the original dataset is imbalanced or heavily skewed. 

```python
#countplt
plt.figure(figsize=(7,6))
sns.countplot(x='Class',data=df,palette='dark')
plt.title('Class Distribution \n (0: no Fraud || 1: Fraud)',fontsize=10)
```

By seeing the distribution we can have an idea how skewed are these features. we can also see further distributions of the other features.

```python
# correlation heatmap
plt.figure(figsize=(20,18))
sns.heatmap(df.corr(),annot=True,linewidth=.5,cmap='Blues')
```
